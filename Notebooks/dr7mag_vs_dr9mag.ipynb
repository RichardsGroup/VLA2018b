```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("quasars_dr7and9.csv")
df.head()
```

```python
z = df[["Z_VI"]].values.flatten()
dr7_mag = df[["MI_Z2"]].values.flatten()
dr9_mag = df[["MI"]].values.flatten()
```

```python
fig = plt.figure(figsize=(10,10))
plt.hist([dr7_mag, dr9_mag], histtype="step", linewidth=4, bins="scott")
plt.xlim(-33, -20)
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter(z, dr7_mag, s=1)
plt.scatter(z, dr9_mag, s=1)
plt.ylim(-32, plt.gca().get_ylim()[1])
```

```python
fig = plt.figure(figsize=(10,10))
plt.hist(dr7_mag-dr9_mag, histtype="step", linewidth=4, bins="scott")
plt.xlim(-1, 1)
plt.show()
```
