```python
import matplotlib.pyplot as plt 
import matplotlib.image as img
import matplotlib.gridspec as gridspec
import richardsplot
import os
```

```python
path = 'NoAnnotations/90as/'
D = os.listdir(path)
for i in D:
    if not i.startswith('.'):    #ignore hidden files/directories
        im = img.imread(path+i)
        plt.imshow(im)
```

```python
fig, axs = plt.subplots(6, 4, figsize=(16,24))

r = 0
c = 0
D = os.listdir(path)
#Need to adjust formatting for our 'odd' number of detections
D.insert(-2, '')


for i in D:
    if not i.startswith('.'):    #ignore hidden files/directories
        if r==5 and c==0:
            axs[r,c].axis('off')
        else:     
            im = img.imread(path+i)
            axs[r,c].imshow(im)

            title = i.split('_')
            axs[r,c].set_title("J"+title[0], fontsize=32.5)
            axs[r,c].axes.xaxis.set_ticklabels([])
            axs[r,c].axes.yaxis.set_ticklabels([])
            axs[r,c].xaxis.set_ticks_position('none')
            axs[r,c].yaxis.set_ticks_position('none')

        c += 1
        c = c%4
        if c==0: r+=1
            
axs[5,3].axis('off')
plt.subplots_adjust(wspace=0.0, hspace=0.145)

"""cbar = fig.colorbar(map1, cax=axins1, orientation="horizontal")
cbar.ax.set_xlabel("W1", fontsize=30)
cbar.ax.xaxis.tick_top()
cbar.ax.xaxis.set_label_position("top")
cbar.ax.tick_params(labelsize=20)"""

plt.tight_layout()
plt.savefig("../PaperPlots/detections_imgGrid.pdf")
```
