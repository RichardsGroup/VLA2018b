I spent most of my week just going through each observation and trying to get as many new images as I could, so there's a good amount of new ones shown in this notebook.  I found a couple of new detections this week by using the scaling tool to reveal the fainter sources that weren't obviously present otherwise.  I have a feeling that I will probably need to go back and re-check some of the earlier images I made and look for some faint sources that I possibly missed.

I also started calculating the radio luminosities for each of our observations using the equation from the Kratzner paper:

<img width="380" alt="Screen Shot 2019-07-19 at 10 06 10 AM" src="https://user-images.githubusercontent.com/52674451/61541243-4f1b5980-aa0d-11e9-8e55-92919e73b797.png">

After applying the equation to about half of the completed images, the log(luminosity) for each observation was in the range of ~30-32erg/s/Hz.  Using the units from the Cozial paper, the luminosities were of the order ~23-25W/Hz
