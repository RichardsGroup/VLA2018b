This week I finished working through all of the observations for our data set.  After finishing, I went back through the images again to check for any faint detections that I may have missed and I actually recovered quite a few more detections/possible detections.  All of the finished images are now stored on github and organized into subdirectories.  

I also calculated the radio luminosities for each of the observations where the image didn't turn out to be distorted and plotted the luminosity against the redshift for each object.  I also made some new plots relating the observations' RMS values to their time spent on source and the number of antennas used in the observation.

```python
import xlrd
file_location = "/Users/trevormccaffrey/Desktop/RMS_Trend_Graphs.xlsx"
workbook = xlrd.open_workbook(file_location)
sheet = workbook.sheet_by_index(0)  #work with first sheet
```

```python
num_antennas = []
time_on_source = []
nt_product = [] #product of each observations time on source and number of antennas
dirty_rms = []
clean_rms = []

nd_redshift = [] #non-detection redshifts
d_redshift = [] #detection redshifts

nd_logL = []
d_logL = []

for n in range(1, sheet.nrows): #17 is number of present-completed images, don't include title
    num_antennas.append(sheet.cell_value(n, 1))
    time_on_source.append(sheet.cell_value(n, 2))
    nt_product.append(sheet.cell_value(n, 3))
    dirty_rms.append(sheet.cell_value(n, 4))
    clean_rms.append(sheet.cell_value(n, 5))
    
    if sheet.cell_value(n, 6) == 0:  #check peak flux value to see if observation is a detection
        nd_redshift.append(sheet.cell_value(n, 8))
        nd_logL.append(sheet.cell_value(n, 12))
    else:
        d_redshift.append(sheet.cell_value(n, 8))
        d_logL.append(sheet.cell_value(n, 12))
```

```python
import matplotlib.pyplot as plt

plt.plot(nt_product, dirty_rms, 'bs')
plt.xlabel('N*t (# antennas and time product)')
plt.ylabel('Dirty RMS (uJy)')
plt.title('Dirty RMS vs N*t')
plt.show()

plt.plot(nt_product, clean_rms, 'rs')
plt.xlabel('N*t (# antennas and time product)')
plt.ylabel('Clean RMS (uJy)')
plt.title('Clean RMS vs N*t')
plt.show()

plt.plot(nd_redshift, nd_logL, 'bo', label = 'Non-Detections')
plt.plot(d_redshift, d_logL, 'ro', label = 'Detections')
plt.xlabel('Redshift')
plt.ylabel('log (Lrad / erg/s/Hz)')
plt.title('Luminosities vs Redshift')
plt.legend(bbox_to_anchor=(1.38, 1))
plt.show()
```

Non-Detections have luminosities less than what is plotted
