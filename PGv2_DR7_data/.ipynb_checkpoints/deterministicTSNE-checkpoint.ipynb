Each target has its own color -- now we want to 'follow' these groups of targets around more runs of t-SNE

General rotation matrix to be applied to a point (x,y) to rotate the projection degree $\theta$ :

\begin{equation*}
R = 
\begin{pmatrix}
\cos\theta & -\sin\theta  \\
\sin\theta & \cos\theta
\end{pmatrix}
\end{equation*}

```python
def plotTSNE(projTSNE2, ax, orientation=False):
    x0 = 0 #take center of circle to be 0,0
    y0 = 0
    x  = projTSNE2[:,0]
    y  = projTSNE2[:,1]
    # Find radial coordinate of each point
    r  = np.sqrt((x - x0)**2 + (y - y0)**2)
    print("'Greenest' point distance from center: {}".format(r[np.argmax(W1)]))
    
    
    if orientation:
        gx = x[np.argmax(W1)] #x-coord of greenest point in new projection
        gy = y[np.argmax(W1)] #y-coord of greenest point
        gpt = np.array([gx,gy])
        norm2 = np.linalg.norm(gpt) #save how far away it is
        func = (gy/gx) #line through origin and greenest pt -- normalization may not be right
        
        #This basically runs through the cases to checkif we need to 'mirror' our data along 'func'
        #Ideally we'd just reflect with projection*-func, but that messes up the scaling
        if gx > 0:
            if y[np.argmin(W1)] > func*x[np.argmin(W1)]:
                for i in range(len(projTSNE2)): projTSNE2[i] = mirror(func, projTSNE2[i]) #mirror the new points along func if needed
        elif gx < 0:
            if y[np.argmin(W1)] < func*x[np.argmin(W1)]:
                for i in range(len(projTSNE2)): projTSNE2[i] = mirror(func, projTSNE2[i])
        print("({},{})".format(gx,gy))

        #Calculate angle between greenest points in each orientation and apply rotation matrix
        print(np.dot(greenestpt,gpt) / (norm1*norm2))
        A = np.dot(greenestpt,gpt) / (norm1*norm2)
        if abs(A) > 1: A=1
        theta = np.arccos(A)
        if gy > m*gx: theta = 2*np.pi - theta #if point above 'slope', want to effectively rotate CW, rather than CCW
        R = np.array(([np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]))
        for i in range(len(projTSNE2)): projTSNE2[i] = np.matmul(R, projTSNE2[i]) #apply rotation matrix to each pt in projection
    
    
    projTSNE2col = projTSNE2[cMask>0]
    projTSNE2radcol = projTSNE2[(cMask>0)&(RLmask)]
    projTSNE2colz = projTSNE2[(cMask>0)&(zmask)]

    ax.scatter(projTSNE2col[:,0], projTSNE2col[:,1], c=W1[cMask>0], cmap="PRGn", edgecolor="None", s=10)
    ax.set_xlabel("Proj1")
    ax.set_ylabel("Proj2")
    #sm = plt.cm.ScalarMappable(cmap="PRGn", norm=plt.Normalize(vmin=0, vmax=1))
    #sm._A = []
    #cbar = plt.colorbar(sm) 
    #cbar.ax.set_ylabel('W1')
    #ax.scatter(projTSNE2radcol[:,0], projTSNE2radcol[:,1], c='w', edgecolor='k', s=30, label="FIRST")
    
    #Color-code targets based on which quadrant they fall in
    ax.scatter(projTSNE2colz[:,0][Q0], projTSNE2colz[:,1][Q0], c=colors[0], edgecolor='w', s=400, marker='*', label="Targets")
    ax.scatter(projTSNE2colz[:,0][Q1], projTSNE2colz[:,1][Q1], c=colors[1], edgecolor='w', s=400, marker='*')
    ax.scatter(projTSNE2colz[:,0][Q2], projTSNE2colz[:,1][Q2], c=colors[2], edgecolor='w', s=400, marker='*')
    ax.scatter(projTSNE2colz[:,0][Q3], projTSNE2colz[:,1][Q3], c=colors[3], edgecolor='w', s=400, marker='*')
    ax.scatter(projTSNE2colz[:,0][Q4], projTSNE2colz[:,1][Q4], c=colors[4], edgecolor='w', s=400, marker='*')
    
    ax.set_xlim(-100,100)
    ax.set_ylim(-100,100)
    ax.legend(loc="upper left")

    circle=plt.Circle((0, 0), r0, color='k', fill=False, linewidth=2)
    ax.add_artist(circle)

    ax.plot([0,0], [r0,100], c='k', linewidth=2)
    ax.plot([0,0], [-r0,-100], c='k', linewidth=2)
    ax.plot([r0,100], [0,0], c='k', linewidth=2)
    ax.plot([-r0,-100], [0,0], c='k', linewidth=2)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,24))
k = 0
for i in range(2):
    for j in range(3):
        plotTSNE(projections1[k], axs[j,i])
        k += 1
plt.savefig("g.pdf")
```

```python
def mirror(m, pt): #derived matrix to mirror across y=mx
    A = (1/(1+m**2)) * np.array(([[(1-m**2), 2*m],
                                  [2*m, (m**2-1)]]))
    
    return np.matmul(A, pt)
```
